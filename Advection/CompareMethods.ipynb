```python
import numpy as np
import matplotlib.pyplot as plt

import scienceplots
plt.style.use(['science', 'notebook', 'grid']) 
plt.rcParams['figure.figsize'] = [10, 10]
```

# 1) FTCS

```python
u_100 = np.load('./results/FTCS_u100.npy')
u_1k = np.load('./results/FTCS_u1000.npy')
u_10k = np.load('./results/FTCS_u10000.npy')
l2norm_100 = np.load('./results/FTCS_l2norm100.npy')
l2norm_1k = np.load('./results/FTCS_l2norm1000.npy')
l2norm_10k = np.load('./results/FTCS_l2norm10000.npy')
```

```python
plt.plot(np.linspace(0, 20, len(u_100)), u_100, label='100 points')
plt.plot(np.linspace(0, 20, len(u_1k)), u_1k, label='1k points')
plt.plot(np.linspace(0, 20, len(u_10k)), u_10k, label='10k points')
plt.xlabel('Time')
plt.ylabel('u(t=20)')
plt.legend()
```

```python
plt.plot(np.linspace(0, 20, len(l2norm_100)), l2norm_100, label='100 points')
plt.plot(np.linspace(0, 20, len(l2norm_1k)), l2norm_1k, label='1k points')
plt.plot(np.linspace(0, 20, len(l2norm_10k)), l2norm_10k, label='10k points')
plt.xlabel('Time')
plt.ylabel('L2norm')
plt.legend()
```

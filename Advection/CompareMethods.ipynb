# 1) Periodic boundary condition

## 1.1) FTCS

```python
u_100 = np.load('./results/FTCS_u100.npy')
u_1k = np.load('./results/FTCS_u1000.npy')
u_10k = np.load('./results/FTCS_u10000.npy')
u_100k = np.load('./results/FTCS_u100000.npy')
l2norm_100 = np.load('./results/FTCS_l2norm100.npy')
l2norm_1k = np.load('./results/FTCS_l2norm1000.npy')
l2norm_10k = np.load('./results/FTCS_l2norm10000.npy')
l2norm_100k = np.load('./results/FTCS_l2norm100000.npy')
```

```python
plt.plot(np.linspace(0, 20, len(u_100)), u_100, label='100 points')
plt.plot(np.linspace(0, 20, len(u_1k)), u_1k, label='1k points')
plt.plot(np.linspace(0, 20, len(u_10k)), u_10k, label='10k points')
plt.plot(np.linspace(0, 20, len(u_100k)), u_100k, label='100k points')
plt.xlabel('Time')
plt.ylabel('u(t=20)')
plt.legend()
```

```python
plt.plot(np.linspace(0, 20, len(l2norm_100)), l2norm_100, label='100 points')
plt.plot(np.linspace(0, 20, len(l2norm_1k)), l2norm_1k, label='1k points')
plt.plot(np.linspace(0, 20, len(l2norm_10k)), l2norm_10k, label='10k points')
plt.plot(np.linspace(0, 20, len(l2norm_100k)), l2norm_100k, label='100k points')

l2_saved=np.genfromtxt('./l2norm_FTCS.dat', comments='#')
plt.plot(l2_saved[:,0], l2_saved[:,1], '+', color='black', label='100 points check')

plt.xlabel('Time')
plt.ylabel('L2norm')
plt.legend()
```

# 2) Outflow boundary condition

```python
l2norm_outflow_laxfriedrichs = np.load('./results/Outflow_LaxFriedrichs_l2norm100000.npy')
l2norm_outflow_leapfrog = np.load('./results/Outflow_Leapfrog_l2norm100000.npy')
l2norm_outflow_laxwendroff = np.load('./results/Outflow_LaxWendroff_l2norm100000.npy')

plt.plot(np.linspace(0, 20, len(l2norm_outflow_laxfriedrichs)), l2norm_outflow_laxfriedrichs, ls='--', color='green', label='Lax-Friedrichs')
plt.plot(np.linspace(0, 20, len(l2norm_outflow_leapfrog)), l2norm_outflow_leapfrog, ls='--', color='deepskyblue', label='Leapfrog')
plt.plot(np.linspace(0, 20, len(l2norm_outflow_laxwendroff)), l2norm_outflow_laxwendroff, ls=':', lw=3, color='red', label='Lax-Wendroff')
plt.yscale('log')
plt.xlabel('Time')
plt.ylabel('L2norm')
plt.legend()
```

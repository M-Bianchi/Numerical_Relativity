```python
import numpy as np
import matplotlib.pyplot as plt

%matplotlib inline
%config InlineBackend.figure_format='retina'
#!pip install git+https://github.com/M-Bianchi/My_matplotlib_style
plt.style.use('mystyle')
```

# 1) Gaussian

```python
Cons_u_100 = np.load('./results/Cons_u100.npy')
Cons_u_1k = np.load('./results/Cons_u1000.npy')
Cons_u_10k = np.load('./results/Cons_u10000.npy')
Cons_u_100k = np.load('./results/Cons_u100000.npy')
Cons_l2norm_100 = np.load('./results/Cons_l2norm100.npy')
Cons_l2norm_1k = np.load('./results/Cons_l2norm1000.npy')
Cons_l2norm_10k = np.load('./results/Cons_l2norm10000.npy')
Cons_l2norm_100k = np.load('./results/Cons_l2norm100000.npy')
```

```python
NONCons_u_100 = np.load('./results/NONCons_u100.npy')
NONCons_u_1k = np.load('./results/NONCons_u1000.npy')
NONCons_u_10k = np.load('./results/NONCons_u10000.npy')
NONCons_u_100k = np.load('./results/NONCons_u100000.npy')
NONCons_l2norm_100 = np.load('./results/NONCons_l2norm100.npy')
NONCons_l2norm_1k = np.load('./results/NONCons_l2norm1000.npy')
NONCons_l2norm_10k = np.load('./results/NONCons_l2norm10000.npy')
NONCons_l2norm_100k = np.load('./results/NONCons_l2norm100000.npy')
```

## 1.1) Function at $t_{\rm final}$ comparison

```python
plt.plot(np.linspace(0, 0.5, len(Cons_u_100)), Cons_u_100, label='Cons, 100 points', ls='-', lw=3, color='lightseagreen', zorder=5)
plt.plot(np.linspace(0, 0.5, len(Cons_u_1k)), Cons_u_1k, label='Cons, 1k points', ls='--', lw=3, color='orange', zorder=4)
plt.plot(np.linspace(0, 0.5, len(Cons_u_10k)), Cons_u_10k, label='Cons, 10k points', ls='-', lw=3, color='royalblue', zorder=3)
plt.plot(np.linspace(0, 0.5, len(NONCons_u_100)), NONCons_u_100, label='NON-Cons, 100 points', ls='-', color='olive', lw=3, zorder=2)
plt.plot(np.linspace(0, 0.5, len(NONCons_u_1k)), NONCons_u_1k, label='NON-Cons, 1k points', ls='--', lw=3, color='black', zorder=1)
plt.plot(np.linspace(0, 0.5, len(NONCons_u_10k)), NONCons_u_10k, label='NON-Cons, 10k points', ls='-', lw=3, color='crimson', zorder=0)
plt.xlabel('Time')
plt.ylabel('u(t=0.5)')
plt.legend()
```

## 1.2) L2norm comparison

```python
plt.plot(np.linspace(0, 0.5, len(Cons_l2norm_100)), Cons_l2norm_100, label='Cons, 100 points', ls='-', lw=3, color='lightseagreen', zorder=5)
plt.plot(np.linspace(0, 0.5, len(Cons_l2norm_1k)), Cons_l2norm_1k, label='Cons, 1k points', ls='--', lw=3, color='orange', zorder=4)
plt.plot(np.linspace(0, 0.5, len(Cons_l2norm_10k)), Cons_l2norm_10k, label='Cons, 10k points', ls='-', lw=3, color='royalblue', zorder=3)
plt.plot(np.linspace(0, 0.5, len(NONCons_l2norm_100)), NONCons_l2norm_100, label='NON-Cons, 100 points', ls='-', color='olive', lw=3, zorder=2)
plt.plot(np.linspace(0, 0.5, len(NONCons_l2norm_1k)), NONCons_l2norm_1k, label='NON-Cons, 1k points', ls='--', lw=3, color='black', zorder=1)
plt.plot(np.linspace(0, 0.5, len(NONCons_l2norm_10k)), NONCons_l2norm_10k, label='NON-Cons, 10k points', ls='-', lw=3, color='crimson', zorder=0)
plt.xlabel('Time')
plt.ylabel('L2norm')
plt.legend()
```

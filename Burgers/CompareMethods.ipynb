```python
import numpy as np
import matplotlib.pyplot as plt

%matplotlib inline
%config InlineBackend.figure_format='retina'
#!pip install git+https://github.com/M-Bianchi/My_matplotlib_style
plt.style.use('mystyle')
plt.rcParams["figure.figsize"] = (8,6)
```

# 1) Gaussian

```python
Cons_u_100 = np.load('./results/Cons_u100.npy')
Cons_u_1k = np.load('./results/Cons_u1000.npy')
Cons_u_10k = np.load('./results/Cons_u10000.npy')
Cons_u_100k = np.load('./results/Cons_u100000.npy')
Cons_l2norm_100 = np.load('./results/Cons_l2norm100.npy')
Cons_l2norm_1k = np.load('./results/Cons_l2norm1000.npy')
Cons_l2norm_10k = np.load('./results/Cons_l2norm10000.npy')
Cons_l2norm_100k = np.load('./results/Cons_l2norm100000.npy')
```

```python
NONCons_u_100 = np.load('./results/NONCons_u100.npy')
NONCons_u_1k = np.load('./results/NONCons_u1000.npy')
NONCons_u_10k = np.load('./results/NONCons_u10000.npy')
NONCons_u_100k = np.load('./results/NONCons_u100000.npy')
NONCons_l2norm_100 = np.load('./results/NONCons_l2norm100.npy')
NONCons_l2norm_1k = np.load('./results/NONCons_l2norm1000.npy')
NONCons_l2norm_10k = np.load('./results/NONCons_l2norm10000.npy')
NONCons_l2norm_100k = np.load('./results/NONCons_l2norm100000.npy')
```

## 1.1) Function at $t_{\rm final}$ comparison

```python
plt.plot(np.linspace(0, 10, len(Cons_u_100)), Cons_u_100, label='Cons, 100 points', ls='-', lw=3, color='lightseagreen')
plt.plot(np.linspace(0, 10, len(Cons_u_10k)), Cons_u_10k, label='Cons, 10k points', ls='-', lw=3, color='royalblue')
plt.plot(np.linspace(0, 10, len(NONCons_u_100)), NONCons_u_100, label='NON-Cons, 100 points', ls='-', color='orange', lw=3)
plt.plot(np.linspace(0, 10, len(NONCons_u_10k)), NONCons_u_10k, label='NON-Cons, 10k points', ls='-', lw=3, color='crimson')
plt.xlim(0,10)
plt.xlabel('x')
plt.ylabel('u(x, t=0.5)')
plt.legend();
```

## 1.2) L2norm comparison

```python
plt.plot(np.linspace(0, 0.5, len(Cons_l2norm_100)), Cons_l2norm_100, label='Cons, 100 points', ls='-', lw=3, color='lightseagreen')
plt.plot(np.linspace(0, 0.5, len(Cons_l2norm_10k)), Cons_l2norm_10k, label='Cons, 10k points', ls='-', lw=3, color='royalblue')
plt.plot(np.linspace(0, 0.5, len(NONCons_l2norm_100)), NONCons_l2norm_100, label='NON-Cons, 100 points', ls='-', color='orange', lw=3)
plt.plot(np.linspace(0, 0.5, len(NONCons_l2norm_10k)), NONCons_l2norm_10k, label='NON-Cons, 10k points', ls='-', lw=3, color='crimson')
plt.xlim(0,0.5)
plt.xlabel('Time')
plt.ylabel('L2norm');
#plt.legend();
```

## 1.3) Time evolution

```python
te_1000 = np.load('./results/Cons_u10000_1000.npy')
te_2000 = np.load('./results/Cons_u10000_2000.npy')
te_3000 = np.load('./results/Cons_u10000_3000.npy')
te_4000 = np.load('./results/Cons_u10000_4000.npy')
te_5000 = np.load('./results/Cons_u10000_5000.npy')
te_6000 = np.load('./results/Cons_u10000_6000.npy')
te_7000 = np.load('./results/Cons_u10000_7000.npy')
te_8000 = np.load('./results/Cons_u10000_8000.npy')
te_9000 = np.load('./results/Cons_u10000_9000.npy')
te_10000 = np.load('./results/Cons_u10000_10000.npy')
```

```python
plt.plot(np.linspace(0, 10, len(te_1000)), te_1000, ls='--', lw=1, color='royalblue', label=f't={0.1*0.5:.2f}')
plt.plot(np.linspace(0, 10, len(te_4000)), te_4000, ls='--', lw=1, color='red', label=f't={0.4*0.5:.2f}')
plt.plot(np.linspace(0, 10, len(te_7000)), te_7000, ls='--', lw=1, color='green', label=f't={0.7*0.5:.2f}')
plt.plot(np.linspace(0, 10, len(te_10000)), te_10000, ls='-', lw=1, color='k', label=f't={1*0.5:.2f}')
plt.xlim(0,10)
plt.xlabel('x')
plt.ylabel('u(x, t)')
plt.legend();
```

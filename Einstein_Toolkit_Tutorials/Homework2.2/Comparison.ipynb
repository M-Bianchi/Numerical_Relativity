```python
import matplotlib.pyplot as plt
import numpy as np

import kuibit.simdir as sd

from kuibit import grid_data as gd

#%matplotlib inline
%config InlineBackend.figure_format='retina'
#!pip install git+https://github.com/M-Bianchi/My_matplotlib_style
plt.style.use('mystyle')
```

```python
timeseries_d2_4 = sd.SimDir("./TOV_example_d2_4/").timeseries
timeseries_d2 = sd.SimDir("./TOV_example_d2/").timeseries
timeseries_d1 = sd.SimDir("./TOV_example_d1/").timeseries
timeseries_d05 = sd.SimDir("./TOV_example_d05/").timeseries
```

```python
vars2D_d2_4 = sd.SimDir("./TOV_example_d2_4/").gf.xy
vars2D_d2 = sd.SimDir("./TOV_example_d2/").gf.xy
vars2D_d1 = sd.SimDir("./TOV_example_d1/").gf.xy
vars2D_d05 = sd.SimDir("./TOV_example_d05/").gf.xy
```

# 1) Density profiles

```python
rho_d2_4 = vars2D_d2_4.fields.rho
rho_d2 = vars2D_d2.fields.rho
rho_d1 = vars2D_d1.fields.rho
rho_d05 = vars2D_d05.fields.rho
```

```python
small_grid = gd.UniformGrid([100, 100], x0=[0, 0], x1=[12,12])

rho_d2_4_small = rho_d2_4.read_on_grid(0, small_grid)
rho_d2_small = rho_d2.read_on_grid(0, small_grid)
rho_d1_small = rho_d1.read_on_grid(0, small_grid)
rho_d05_small = rho_d05.read_on_grid(0, small_grid)
```

```python
fig, axes = plt.subplots(2, 2, figsize=(12,12))

ax = axes[0,0]
ax.set_box_aspect(1)
ax.set_title('$dx=dy=dz=2.4$')
ax.pcolormesh(*rho_d2_4_small.coordinates_meshgrid(), np.log10(rho_d2_4_small.data_xyz), vmin=-10, vmax=-3, cmap=plt.cm.Blues, zorder=0)

ax = axes[0,1]
ax.set_box_aspect(1)
ax.set_title('$dx=dy=dz=2.0$')
ax.pcolormesh(*rho_d2_small.coordinates_meshgrid(), np.log10(rho_d2_small.data_xyz), vmin=-10, vmax=-3, cmap=plt.cm.Blues, zorder=0)

ax = axes[1,0]
ax.set_box_aspect(1)
ax.set_title('$dx=dy=dz=1.0$')
ax.pcolormesh(*rho_d1_small.coordinates_meshgrid(), np.log10(rho_d1_small.data_xyz), vmin=-10, vmax=-3, cmap=plt.cm.Blues, zorder=0)

ax = axes[1,1]
cax = fig.add_axes([0.93, 0.11, 0.03, 0.77])
ax.set_box_aspect(1)
ax.set_title('$dx=dy=dz=0.5$')
cf = ax.pcolormesh(*rho_d05_small.coordinates_meshgrid(), np.log10(rho_d05_small.data_xyz), vmin=-10, vmax=-3, cmap=plt.cm.Blues, zorder=0)
plt.colorbar(cf, cax=cax, label='$\\log_{10}\\rho(t=0)$');
```

```python
rho_d2_4_small_final = rho_d2_4.read_on_grid(rho_d2_4.iterations[-1], small_grid)
rho_d2_small_final = rho_d2.read_on_grid(rho_d2.iterations[-1], small_grid)
rho_d1_small_final = rho_d1.read_on_grid(rho_d1.iterations[-1], small_grid)
rho_d05_small_final = rho_d05.read_on_grid(rho_d05.iterations[-1], small_grid)

fig, axes = plt.subplots(2, 2, figsize=(12,12))

ax = axes[0,0]
ax.set_box_aspect(1)
ax.set_title('$dx=dy=dz=2.4$')
ax.pcolormesh(*rho_d2_4_small_final.coordinates_meshgrid(), np.log10(rho_d2_4_small_final.data_xyz), vmin=-10, vmax=-3, cmap=plt.cm.Blues, zorder=0)

ax = axes[0,1]
ax.set_box_aspect(1)
ax.set_title('$dx=dy=dz=2.0$')
ax.pcolormesh(*rho_d2_small_final.coordinates_meshgrid(), np.log10(rho_d2_small_final.data_xyz), vmin=-10, vmax=-3, cmap=plt.cm.Blues, zorder=0)

ax = axes[1,0]
ax.set_box_aspect(1)
ax.set_title('$dx=dy=dz=1.0$')
ax.pcolormesh(*rho_d1_small_final.coordinates_meshgrid(), np.log10(rho_d1_small_final.data_xyz), vmin=-10, vmax=-3, cmap=plt.cm.Blues, zorder=0)

ax = axes[1,1]
cax = fig.add_axes([0.93, 0.11, 0.03, 0.77])
ax.set_box_aspect(1)
ax.set_title('$dx=dy=dz=0.5$')
cf = ax.pcolormesh(*rho_d05_small_final.coordinates_meshgrid(), np.log10(rho_d05_small_final.data_xyz), vmin=-10, vmax=-3, cmap=plt.cm.Blues, zorder=0)
plt.colorbar(cf, cax=cax, label='$\\log_{10}\\rho(t=400)$');
```

```python
fig, axes = plt.subplots(2, 4, figsize=(18,8))

ax = axes[0,0]
ax.set_box_aspect(1)
ax.set_title('$dx=dy=dz=2.4$')
ax.pcolormesh(*rho_d2_4_small.coordinates_meshgrid(), np.log10(rho_d2_4_small.data_xyz), vmin=-10, vmax=-3, cmap=plt.cm.Blues, zorder=0)
ax.set_ylabel('$t=0$', fontsize=30)

ax = axes[0,1]
ax.set_box_aspect(1)
ax.set_title('$dx=dy=dz=2.0$')
ax.pcolormesh(*rho_d2_small.coordinates_meshgrid(), np.log10(rho_d2_small.data_xyz), vmin=-10, vmax=-3, cmap=plt.cm.Blues, zorder=0)

ax = axes[0,2]
ax.set_box_aspect(1)
ax.set_title('$dx=dy=dz=1.0$')
ax.pcolormesh(*rho_d1_small.coordinates_meshgrid(), np.log10(rho_d1_small.data_xyz), vmin=-10, vmax=-3, cmap=plt.cm.Blues, zorder=0)

ax = axes[0,3]
ax.set_box_aspect(1)
ax.set_title('$dx=dy=dz=0.5$')
cf = ax.pcolormesh(*rho_d05_small.coordinates_meshgrid(), np.log10(rho_d05_small.data_xyz), vmin=-10, vmax=-3, cmap=plt.cm.Blues, zorder=0)

ax = axes[1,0]
ax.set_box_aspect(1)
ax.pcolormesh(*rho_d2_4_small_final.coordinates_meshgrid(), np.log10(rho_d2_4_small_final.data_xyz), vmin=-10, vmax=-3, cmap=plt.cm.Blues, zorder=0)
ax.set_ylabel('$t=400$', fontsize=30)

ax = axes[1,1]
ax.set_box_aspect(1)
ax.pcolormesh(*rho_d2_small_final.coordinates_meshgrid(), np.log10(rho_d2_small_final.data_xyz), vmin=-10, vmax=-3, cmap=plt.cm.Blues, zorder=0)

ax = axes[1,2]
ax.set_box_aspect(1)
ax.pcolormesh(*rho_d1_small_final.coordinates_meshgrid(), np.log10(rho_d1_small_final.data_xyz), vmin=-10, vmax=-3, cmap=plt.cm.Blues, zorder=0)

ax = axes[1,3]
cax = fig.add_axes([0.91, 0.11, 0.015, 0.77])
ax.set_box_aspect(1)
cf = ax.pcolormesh(*rho_d05_small_final.coordinates_meshgrid(), np.log10(rho_d05_small_final.data_xyz), vmin=-10, vmax=-3, cmap=plt.cm.Blues, zorder=0)
plt.colorbar(cf, cax=cax, label='$\\log_{10}\\rho(t)$');
```

# 2) Time series of the maximum central rest-mass density

```python
plt.plot(timeseries_d05.maximum.fields.rho, c='red', label="$dx=dy=dz=0.5$")
plt.plot(timeseries_d1.maximum.fields.rho, ls=':', c='black', label="$dx=dy=dz=1.0$")
plt.plot(timeseries_d2.maximum.fields.rho, ls='--', c='royalblue', label="$dx=dy=dz=2.0$")
plt.plot(timeseries_d2_4.maximum.fields.rho, ls='-.', c='orange', label="$dx=dy=dz=2.4$")
plt.axhline(0.00128, c='gray', ls='--', lw=1, label='$\\rho_\\text{center}\\left(t=0\\right)$', zorder=0)
plt.xlim(0,400)
plt.xlabel('$t$')
plt.ylabel('$\\rho_\\text{center}$')
plt.legend();
```

# 3) Time series of the maximum central pressure

```python
plt.plot(timeseries_d05.maximum.fields.press, c='red', label="$dx=dy=dz=0.5$")
plt.plot(timeseries_d1.maximum.fields.press, ls=':', c='black', label="$dx=dy=dz=1.0$")
plt.plot(timeseries_d2.maximum.fields.press, ls='--', c='royalblue', label="$dx=dy=dz=2.0$")
plt.plot(timeseries_d2_4.maximum.fields.press, ls='-.', c='orange', label="$dx=dy=dz=2.4$")
plt.xlim(0,400)
plt.xlabel('$t$')
plt.ylabel('$P_\\text{center}$')
plt.legend();
```
